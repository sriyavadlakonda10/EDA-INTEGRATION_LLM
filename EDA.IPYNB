```python
import ollama
import pandas as pd
import subprocess

# Function to check if the model is available
def check_model_exists(model_name):
    try:
        models = subprocess.run(["ollama", "list"], capture_output=True, text=True)
        return model_name in models.stdout
    except Exception as e:
        return False

def generate_insights(df_summary, model="mistral"):
    if not check_model_exists(model):
        return f"Error: Model '{model}' not found. Run 'ollama pull {model}' in the terminal."

    try:
        prompt = f"Analyze the dataset summary and provide insights:\n\n{df_summary}"
        response = ollama.chat(model=model, messages=[{"role": "user", "content": prompt}])
        return response['message']['content']
    except Exception as e:
        return f"Error: {e}"

# Sample DataFrame for testing
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
})

# Generate AI Insights
summary = df.describe().to_string()
insights = generate_insights(summary)
print("\n🔹 AI-Generated Insights:\n", insights)
```

```python
import gradio as gr

def eda_analysis(file):
    df = pd.read_csv(file.name)
    summary = df.describe().to_string()
    insights = generate_insights(summary)
    return insights

# Create Web Interface
demo = gr.Interface(fn=eda_analysis, inputs="file", outputs="text", title="AI-Powered EDA with Mistral")

# Launch App
demo.launch(share=True)  # Use share=True for Google Colab
```
